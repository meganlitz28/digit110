3b. Prepare a code block that depends on running the previous code block and then reads from each directory of files and calculates what's needed to indicate **lexical diversity**.
    Running the cell that you prepare should reveal the lexical diversity of each set of files. *(3 pts)*

```python
import os

corpus_path_1 = 'season 5' 
corpus_path_2 = 'season 8'

def calculate_lexical_diversity(directory_path):
    all_text = ""
    
    # 1. Loop through the directory and read all files
    for filename in os.listdir(directory_path):
        if filename.endswith(".txt"):
            with open(os.path.join(directory_path, filename), 'r', encoding='utf-8') as f:
                all_text += f.read().lower() + " "
    
    # 2. Tokenize the combined text (simple split for diversity check)
    tokens = nltk.word_tokenize(all_text)
    
    # 3. Filter out punctuation to get a cleaner count
    words = [word for word in tokens if word.isalnum()]
    
    # 4. Calculate Diversity: Unique Words / Total Words
    total_tokens = len(words)
    unique_types = len(set(words))
    
    if total_tokens == 0:
        return 0
    
    diversity_score = unique_types / total_tokens
    return diversity_score, total_tokens, unique_types

score1, tokens1, types1 = calculate_lexical_diversity(corpus_path_1)
score2, tokens2, types2 = calculate_lexical_diversity(corpus_path_2)

print(f"Results for {corpus_path_1}:")
print(f" - Total Words: {tokens1}")
print(f" - Unique Words: {types1}")
print(f" - Lexical Diversity: {score1:.4f}\n")

print(f"Results for {corpus_path_2}:")
print(f" - Total Words: {tokens2}")
print(f" - Unique Words: {types2}")
print(f" - Lexical Diversity: {score2:.4f}")
```

```python
def extract_pos_tokens(directory_path, pos_type="ADJ"):
   
    pos_tokens = []
    
    for filename in os.listdir(directory_path):
        if filename.endswith(".txt"):
            with open(os.path.join(directory_path, filename), 'r', encoding='utf-8') as f:
                text = f.read()
                
                doc = nlp(text)
                
                for token in doc:
                    if token.pos_ == pos_type and not token.is_punct and not token.is_space:
                        pos_tokens.append(token.text.lower())
    
    return pos_tokens

# Selecting ADJECTIVES from Season 5 and Season 8 for analysis
s5_adjectives = extract_pos_tokens(corpus_path_1, pos_type="ADJ")
s8_adjectives = extract_pos_tokens(corpus_path_2, pos_type="ADJ")

# Preview the first 10 adjectives found in Season 5
print(f"Sample adjectives from {corpus_path_1}: {s5_adjectives[:10]}")
print(f"Total adjectives found: {len(s5_adjectives)}")
```

6. Develop the code in the previous block by copying it into the next code block, and here:
   * Develop code to find out about the top 15 most ambiguous words in your adjective or verb token list, and
   * Write your output to a neatly formatted text file for reading. The text output file should provide a clear heading to provide a title for each corpus you're comparing. *(5 pts)*

```python
import spacy
from collections import Counter

def analyze_ambiguity(directory_path, target_pos="ADJ"):
    all_text = ""
    for filename in os.listdir(directory_path):
        if filename.endswith(".txt"):
            with open(os.path.join(directory_path, filename), 'r', encoding='utf-8') as f:
                all_text += f.read() + " "
    
    doc = nlp(all_text)
    
    word_pos_map = {}
    for token in doc:
        if not token.is_punct and not token.is_space:
            lemma = token.lemma_.lower()
            pos = token.pos_
            if lemma not in word_pos_map:
                word_pos_map[lemma] = set()
            word_pos_map[lemma].add(pos)

    ambiguous_data = []
    for word, pos_set in word_pos_map.items():
        if len(pos_set) > 1:
            ambiguous_data.append((word, len(pos_set), list(pos_set)))
            
    # Sort by the most tags (highest ambiguity)
    ambiguous_data.sort(key=lambda x: x[1], reverse=True)
    return ambiguous_data[:15]

# 2. Run analysis for both corpora
ambig1 = analyze_ambiguity(corpus_path_1)
ambig2 = analyze_ambiguity(corpus_path_2)

with open("ambiguity_results.txt", "w", encoding="utf-8") as out:
    out.write("ARTHUR TV SHOW: AMBIGUITY ANALYSIS\n")
    out.write("="*35 + "\n\n")
    
    out.write(f"TOP 15 AMBIGUOUS WORDS IN {corpus_path_1.upper()}:\n")
    for word, count, tags in ambig1:
        out.write(f"{word:15} | Tags: {count} ({', '.join(tags)})\n")
        
    out.write("\n" + "="*35 + "\n\n")
    
    out.write(f"TOP 15 AMBIGUOUS WORDS IN {corpus_path_2.upper()}:\n")
    for word, count, tags in ambig2:
        out.write(f"{word:15} | Tags: {count} ({', '.join(tags)})\n")

print("Analysis complete! Results saved to 'ambiguity_results.txt'.")
```
